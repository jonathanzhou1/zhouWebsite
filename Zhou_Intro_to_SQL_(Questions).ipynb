```python
#Q13: show the 10 most expensive items in stock
%sql select name, price from items where qty > 0 order by 2 desc limit 10;
```

```python
#Q14: show a list of all item names and their category number. If an item is in multiple categories, show the item once for each category number.
%sql select name, cid as category_number from items inner join item_category on items.id = item_category.pid;
```

```python
#Q15: show a list of all item names, their category name, and their subcategory name. If an item is in multiple categories, show the item once for each category.
%sql select name, main, sub from items i inner join item_category ic on i.id = ic.pid inner join categories c on ic.cid = c.id;
```

```python
#Q16: For each category, show the number of subcategories
%sql select main, count(sub) from categories group by main;
```

```python
#Q17: For each uid in orders, show the number of orders made by that user uid
%sql select distinct user, count(qty) as total_orders from orders o inner join order_items oi on o.id = oi.oid group by oid;
```

```python
#Q18: for each item, show the number of times it has been ordered
%sql select name, sum(oi.qty) as times_ordered from items i inner join order_items oi on i.id = oi.pid group by name; 
```

```python
#Q19: what is the most popular item in the store?
%sql select name, sum(oi.qty) as times_ordered from items i inner join order_items oi on i.id = oi.pid group by name order by 2 desc limit 1; 
```

```python
#Q20: show all items that have been ordered more than 50 times
%sql select name, sum(oi.qty) as times_ordered from items i inner join order_items oi on i.id = oi.pid group by name having sum(oi.qty) > 50;
```

```python
#Q21: show a list of all items ordered in June 2021 that are currently out of stock
%sql select name, time, i.qty from items i inner join order_items oi on i.id = oi.pid inner join orders o on oi.oid = o.id where i.qty = 0 and time like '2021-06%';
```

```python
#Q22: show the number of orders placed every month starting from January 2021
%sql select sum(qty) as total_orders, MONTH(time) from orders o inner join order_items oi on o.id = oi.oid where time > '2021' group by MONTH(time);
```

```python
#Q23: how much money has been spent at the store?
%sql select sum(total)/100 as total_money from orders;
```

```python
#Q24: how much tip has been made?
%sql select sum(tip)/100 as total_tips from orders;
```

```python
#Q25: who has the fastest delivered orders on average? what is their average delivery time?
%sql select user_name, AVG((delivered - time)/100) as delivery_time from orders group by id order by 2 limit 1;
```

```python
#Q26: what is the average delivery time in each year-month? hopefully the delivery time is improving!
%sql select YEAR(time) as year, MONTH(time) as month, AVG((delivered - time)/100) as delivery_time from orders group by YEAR(time), MONTH(time);
```

```python
#Q27: how many items have been ordered from each main category?
%sql select distinct(main) as main_category, sum(oi.qty) as total_orders from orders o inner join order_items oi on o.id = oi.oid inner join items i on oi.pid = i.id inner join item_category ic on i.id = ic.pid inner join categories c on ic.cid = c.id group by main;
```

```python
  #Q28: what are the top three most popular main categories?
%sql select distinct(main) as main_category, sum(oi.qty) as total_orders from orders o inner join order_items oi on o.id = oi.oid inner join items i on oi.pid = i.id inner join item_category ic on i.id = ic.pid inner join categories c on ic.cid = c.id group by main order by 2 desc limit 3;
```

```python
#Q29: what are the top five most popular subcategories?
%sql select distinct(sub) as sub_category, sum(oi.qty) as total_orders from orders o inner join order_items oi on o.id = oi.oid inner join items i on oi.pid = i.id inner join item_category ic on i.id = ic.pid inner join categories c on ic.cid = c.id group by sub order by 2 desc limit 5;
```

```python
#Q30: how much money has been spent on items from each subcategory?
%sql select distinct(sub) as sub_category, sum(o.total/100) as money_spent from orders o inner join order_items oi on o.id = oi.oid inner join items i on oi.pid = i.id inner join item_category ic on i.id = ic.pid inner join categories c on ic.cid = c.id group by sub;
```

```python
#Q31: which hour is the most popular for the store?
%sql select HOUR(time) as hour, COUNT(total) as count from orders group by HOUR(time) order by 2 desc limit 1;
```

```python
#Q32: for each user, what was the first time they placed an order?
%sql select user, MIN(time) from orders group by user;
```
